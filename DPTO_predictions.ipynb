```python
import joblib
import pandas as pd
import shap
```

```python
features = pd.read_csv("./data/ML_Models/selected_variables.csv")['Feature'].tolist()
```

```python
data_df = pd.read_csv("./data/DPTO_INFORMATION_SD_VO.csv")
```

```python
rfc = joblib.load('./data/ML_Models/randomForestClassifier-v1.0.joblib')
```

```python
data_feats = data_df.loc[:,features]
data_feats
```

```python
predictions = rfc.predict(data_feats)
```

```python
predictions
```

```python
rfc.best_estimator_
```

```python
def model(X):
    return rfc.best_estimator_["classifier"].predict_proba(X)[:,1]

explainer = shap.Explainer(model,data_feats)
shap_values = explainer.shap_values(data_feats)
```

```python
shap_df = pd.DataFrame(shap_values)
shap_df = shap_df.rename(columns=dict(zip(shap_df.columns, data_feats.columns)))
shap_df
```
