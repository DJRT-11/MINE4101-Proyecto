```python
Alfa = Alfa.drop(columns="DEPARTAMENTO")
Edad = Edad.drop(columns="DEPARTAMENTO")
Educacion = Educacion.drop(columns="DEPARTAMENTO")
Etnia = Etnia.drop(columns="DEPARTAMENTO")
Estado_Civil = Estado_Civil.drop(columns="DEPARTAMENTO")
Estrato = Estrato.drop(columns="DEPARTAMENTO")
Sanitario = Sanitario.drop(columns="DEPARTAMENTO")
Servicios = Servicios.drop(columns="DEPARTAMENTO")
Sexo = Sexo.drop(columns="DEPARTAMENTO")
Tipos_de_Hogar = Tipos_de_Hogar.drop(columns="DEPARTAMENTO")

Alfa = Alfa.rename(columns={"CODIGO":"DPTO_CCDGO"})
Edad = Edad.rename(columns={"CODIGO":"DPTO_CCDGO"})
Educacion = Educacion.rename(columns={"CODIGO":"DPTO_CCDGO"})
Etnia = Etnia.rename(columns={"CODIGO":"DPTO_CCDGO"})
Estado_Civil = Estado_Civil.rename(columns={"CODIGO":"DPTO_CCDGO"})
Estrato = Estrato.rename(columns={"CODIGO":"DPTO_CCDGO"})
Sanitario = Sanitario.rename(columns={"CODIGO":"DPTO_CCDGO"})
Servicios = Servicios.rename(columns={"CODIGO":"DPTO_CCDGO"})
Sexo = Sexo.rename(columns={"CODIGO":"DPTO_CCDGO"})
Tipos_de_Hogar = Tipos_de_Hogar.rename(columns={"CODIGO":"DPTO_CCDGO"})
```

```python
Alfa = Alfa.rename(columns={"CODIGO":"DPTO_CCDGO"})
Edad = Edad.rename(columns={"CODIGO":"DPTO_CCDGO"})
Educacion = Educacion.rename(columns={"CODIGO":"DPTO_CCDGO"})
Etnia = Etnia.rename(columns={"CODIGO":"DPTO_CCDGO"})
Estado_Civil = Estado_Civil.rename(columns={"CODIGO":"DPTO_CCDGO"})
Estrato = Estrato.rename(columns={"CODIGO":"DPTO_CCDGO"})
Sanitario = Sanitario.rename(columns={"CODIGO":"DPTO_CCDGO"})
Servicios = Servicios.rename(columns={"CODIGO":"DPTO_CCDGO"})
Sexo = Sexo.rename(columns={"CODIGO":"DPTO_CCDGO"})
Tipos_de_Hogar = Tipos_de_Hogar.rename(columns={"CODIGO":"DPTO_CCDGO"})

TOTAL = TOTAL.merge(Alfa,on="DPTO_CCDGO",how="left")
TOTAL = TOTAL.merge(Edad,on="DPTO_CCDGO",how="left")
TOTAL = TOTAL.merge(Educacion,on="DPTO_CCDGO",how="left")
TOTAL = TOTAL.merge(Etnia,on="DPTO_CCDGO",how="left")
TOTAL = TOTAL.merge(Estado_Civil,on="DPTO_CCDGO",how="left")
TOTAL = TOTAL.merge(Estado_Civil,on="DPTO_CCDGO",how="left")
TOTAL = TOTAL.merge(Sanitario,on="DPTO_CCDGO",how="left")
TOTAL = TOTAL.merge(Servicios,on="DPTO_CCDGO",how="left")
TOTAL = TOTAL.merge(Sexo,on="DPTO_CCDGO",how="left")
TOTAL = TOTAL.merge(Tipos_de_Hogar,on="DPTO_CCDGO",how="left")
```
